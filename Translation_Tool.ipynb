```python
# Install if not already installed
!pip install deep-translator gTTS ipywidgets

from deep_translator import GoogleTranslator
import ipywidgets as widgets
from IPython.display import display, Audio
from gtts import gTTS
import tempfile
import os

# -----------------------------
# Functions
# -----------------------------
def translate_text(_):
    src = source_lang.value
    dest = target_lang.value
    text = input_text.value.strip()
    if not text:
        output_text.value = "‚ùå Please enter some text."
        return
    try:
        translated = GoogleTranslator(source=src, target=dest).translate(text)
        output_text.value = translated
    except Exception as e:
        output_text.value = f"‚ùå Error: {str(e)}"

def speak_text(_):
    if not output_text.value.strip():
        return
    tts = gTTS(text=output_text.value, lang=target_lang.value)
    with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as f:
        tts.save(f.name)
        display(Audio(filename=f.name, autoplay=True))
        os.remove(f.name)

def copy_text(_):
    if output_text.value.strip():
        print("üìã Copied Output:", output_text.value)

# -----------------------------
# Widgets
# -----------------------------
input_text = widgets.Textarea(
    placeholder="Enter text to translate...",
    description="Input:",
    layout=widgets.Layout(width="100%", height="80px")
)

source_lang = widgets.Dropdown(
    options=["auto", "en", "fr", "de", "hi", "es", "ta"],
    value="en",
    description="From:"
)

target_lang = widgets.Dropdown(
    options=["en", "fr", "de", "hi", "es", "ta"],
    value="fr",
    description="To:"
)

translate_button = widgets.Button(
    description="Translate",
    button_style="success"
)
translate_button.on_click(translate_text)

output_text = widgets.Textarea(
    placeholder="Translation will appear here...",
    description="Output:",
    layout=widgets.Layout(width="100%", height="80px")
)

tts_button = widgets.Button(description="Speak Output", button_style="info")
tts_button.on_click(speak_text)

copy_button = widgets.Button(description="Copy Output", button_style="warning")
copy_button.on_click(copy_text)

# -----------------------------
# Display UI
# -----------------------------
display(widgets.VBox([
    input_text,
    widgets.HBox([source_lang, target_lang]),
    translate_button,
    output_text,
    widgets.HBox([tts_button, copy_button])
]))
```
